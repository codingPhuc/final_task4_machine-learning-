```python
from keras.datasets import mnist
from sklearn.metrics import accuracy_score
import joblib
import numpy as np
import matplotlib.pyplot as plt 

(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

decision tree

```python
import random
import matplotlib.pyplot as plt
from sklearn import tree 
# reshape the the decision tree
X,y  = train_X.reshape((len(train_X), -1)) , train_y 
X_T,y_T  = test_X.reshape((len(test_X), -1)) , test_y
# Fit the training data into the DecisionTreeClassifier
clf  = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python

train_acc_clf = accuracy_score(y, clf.predict(X))
test_acc_clf = accuracy_score(y_T, clf.predict(X_T))
print(f"Training accuracy: {train_acc_clf}")
print(f"Testing accuracy: {test_acc_clf}")
```

```python
joblib.dump(clf, 'decision_tree_model.joblib')
loaded_model = joblib.load('decision_tree_model.joblib')

sample_1,sample_2,sample_3,sample_4,sample_5 = np.array_split(X_T, 5)
test_1,test_2,test_3,test_4,test_5 = np.array_split(y_T, 5)
predictions =loaded_model.predict(sample_1)
print(predictions)
print(test_1)
predictions =loaded_model.predict(sample_2)
print(predictions)
print(test_2)
predictions =loaded_model.predict(sample_3)
print(predictions)
print(test_3)
predictions =loaded_model.predict(sample_4)
print(predictions)
print(test_4)
predictions =loaded_model.predict(sample_5)
print(predictions)
print(test_5)
```

```python
tree.plot_tree(loaded_model)
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(X, y).predict(X_T)
```

```python
train_acc_gnb = accuracy_score(y, gnb.predict(X))
test_acc_gnb = accuracy_score(y_T, gnb.predict(X_T))
print(f"Training accuracy: {train_acc_gnb}")
print(f"Testing accuracy: {test_acc_gnb}")
```

```python
joblib.dump(clf, 'Naive_Bayes.joblib')
loaded_model = joblib.load('Naive_Bayes.joblib')

sample_1,sample_2,sample_3,sample_4,sample_5 = np.array_split(X_T, 5)
test_1,test_2,test_3,test_4,test_5 = np.array_split(y_T, 5)
predictions =loaded_model.predict(sample_1)
print(predictions)
print(test_1)
predictions =loaded_model.predict(sample_2)
print(predictions)
print(test_2)
predictions =loaded_model.predict(sample_3)
print(predictions)
print(test_3)
predictions =loaded_model.predict(sample_4)
print(predictions)
print(test_4)
predictions =loaded_model.predict(sample_5)
print(predictions)
print(test_5)
```

KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier()
neigh.fit(X, y)
```

```python
joblib.dump(neigh, 'neigh.joblib')
loaded_model = joblib.load('neigh.joblib')

sample_1,sample_2,sample_3,sample_4,sample_5 = np.array_split(X_T, 5)
test_1,test_2,test_3,test_4,test_5 = np.array_split(y_T, 5)
predictions =loaded_model.predict(sample_1)
print(predictions)
print(test_1)
predictions =loaded_model.predict(sample_2)
print(predictions)
print(test_2)
predictions =loaded_model.predict(sample_3)
print(predictions)
print(test_3)
predictions =loaded_model.predict(sample_4)
print(predictions)
print(test_4)
predictions =loaded_model.predict(sample_5)
print(predictions)
print(test_5)
```

```python
train_acc_neigh = accuracy_score(y, neigh.predict(X))
test_acc_neigh = accuracy_score(y_T, neigh.predict(X_T))
print(f"Training accuracy: {train_acc_neigh}")
print(f"Testing accuracy: {test_acc_neigh}")
```

```python


# Define Data

ML_MODEL = ['Decision Tree','Nave Bayes','k-NN model']
train_data= [train_acc_clf, train_acc_gnb , train_acc_neigh]
test_data = [test_acc_clf,test_acc_gnb,test_acc_neigh]

x_axis = np.arange(len(ML_MODEL))

plt.xticks(x_axis, ML_MODEL)
plt.bar(x_axis -0.2, train_data, width=0.4, label = 'train_data',color='red')
plt.bar(x_axis +0.2, test_data, width=0.4, label = 'test_data',color='blue')
plt.legend()
plt.title('machine learnning')
plt.show()
```
